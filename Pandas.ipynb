```python
import pandas as pd
import numpy as np
```

```python
index = pd.MultiIndex.from_product([range(100), range(10)], names=['A', 'B'])
df = pd.DataFrame({'C': np.random.randn(1000)}, index=index)
```

```python
df.head(20)
```

```python
df['C'] += index.get_level_values(1)
```

```python
df.head(20)
```

```python
df2 = df.unstack(1)
```

```python
df2.plot(kind='box', figsize=(8,6))
```
